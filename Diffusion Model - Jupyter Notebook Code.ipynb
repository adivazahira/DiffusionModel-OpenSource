```python
import gradio as gr
from diffusers import StableDiffusionPipeline
import torch

model_id = "CompVis/stable-diffusion-v1-4"
pipe = StableDiffusionPipeline.from_pretrained(model_id)
pipe.to("cuda" if torch.cuda.is_available() else "cpu")

def text2image(prompt):
    # Generate the image
    with torch.no_grad():  # Ensure no gradients are computed to save memory and computations
        generation = pipe(prompt=prompt)

    # Dynamically obtain the key for the generated images
    # This assumes that the pipeline returns a dictionary with one key containing the images
    image_key = next(iter(generation.keys()))
    generated_images = generation[image_key]

    # Return the first image in the batch
    return generated_images[0]

# Create Gradio interface
iface = gr.Interface(fn=text2image, inputs="text", outputs="image", title="Text-to-Image Generator")

# Launch the interface
iface.launch(debug=True)
iface.launch(share=True)
```

#**Image-to-Text Generator**

```python
import gradio as gr
from transformers import BlipProcessor, BlipForConditionalGeneration
from PIL import Image

# Initialize the model and processor
processor = BlipProcessor.from_pretrained("Salesforce/blip-image-captioning-large")
model = BlipForConditionalGeneration.from_pretrained("Salesforce/blip-image-captioning-large")

def image_to_text(image):
    # Process the image
    inputs = processor(text=None, images=image, return_tensors="pt", padding=True)

    # Generate caption
    outputs = model.generate(**inputs)
    caption = processor.decode(outputs[0], skip_special_tokens=True)

    return caption

# Create the Gradio interface
iface = gr.Interface(
    fn=image_to_text,
    inputs=gr.Image(type="pil"),  # Changed from `gr.inputs.Image`
    outputs="text",
    title="Image to Text Generation"
)

# Launch the interface
iface.launch(debug=True)
```
